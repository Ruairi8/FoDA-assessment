### Exercise 2: Show that the difference between the standard deviation calculations is greatest for small sample sizes.

When samples are taken from a population, the sample means can vary but there can only be one population mean. 

Bessel's Correction formula: <br><center>$ n / n - 1 $<center>

If we let n = 2: $\Rightarrow  2 / 2 - 1 = 2.$ <br>
$n = 5 \Rightarrow  5 / 5 - 1 = 1.25$. $n = 100  \Rightarrow  101 / 101 - 1 = 1.01$ <br>
$n = 801 \Rightarrow 801 / 801 - 1 = 1.00125$.

It is evident that Bessel's Correction rapidly approaches 1 for increasing length, n, of a sample from a population. Bessel's Correction is multiplied to the variance of a sample, and when a variance is multiplied by a number very close to 1.0, that variance will change very negligibly.

```python
import numpy as np
x = np.random.normal(7, 2, (600, 3))
y = np.random.normal(7, 2, (600, 100))
```

```python
stdevs_x = x.std(axis=1)
stdevs_x
```

```python
stdevs_y = y.std(axis=1)
stdevs_y
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 6))
plt.hist(stdevs_x, bins=100)
plt.axvline(x=2.0, color='red')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
plt.hist(stdevs_y, bins=100)
plt.axvline(x=2.0, color='red')
```
